```python
# Generate a list of all possible starting two numbers pairs
listA = [(i, j) for i in range(10) for j in range(10)]

# Define a function to generate a sequence of consecutive pairs
def generate_consecutive_pairs(start_pair):
    sequence = [start_pair]
    a, b = start_pair
    while True:
        next_number = (a + b) % 10
        next_pair = (b, next_number)
        if next_pair in sequence:
            break
        sequence.append(next_pair)
        a, b = next_pair
    return sequence

# Initialize a list to store the starting pairs of each unique bracelet
distinct_bracelet_starts = []

#  Loop until listA is empty
while listA:
    # Take the first pair from listA as the starting pair
    start_pair = listA[0]
    distinct_bracelet_starts.append(start_pair)
    
    # Generate the sequence of consecutive pairs
    consecutive_pairs = generate_consecutive_pairs(start_pair)

    # Remove these consecutive pairs from listA
    listA = [pair for pair in listA if pair not in consecutive_pairs]

    # print the current starting pair and its generated sequence
    print(f"Start pair: {start_pair}, Generated sequence: {consecutive_pairs}")

# When listA is empty, the program ends
print("No more pairs left in listA.")


# Counting different bracelets
num_distinct_bracelets = len(distinct_bracelet_starts)
print(f"Total number of different bracelets: {num_distinct_bracelets}")

```
